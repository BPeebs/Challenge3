```python
#Calculate arbitrage spread on January 15, 2018
arbitrage_spread_early = bitstamp_sliced.loc['2018-01-15']-coinbase_sliced.loc['2018-01-15']
arbitrage_spread_early.describe()
```

```python
#calculate spread return greater than 0 on January 15
spread_return_early=arbitrage_spread_early[arbitrage_spread_early>0]/coinbase_sliced.loc["2018-01-15"]
spread_return_early.head()
```

```python
#Calculate arbitrage spread on February 24, 2018
arbitrage_spread_middle = coinbase_sliced.loc['2018-2-24']-bitstamp_sliced.loc['2018-02-24']
arbitrage_spread_middle.describe()
```

```python
#calculate spreads greater than 0 on Feb 24
spread_return_middle=arbitrage_spread_middle[arbitrage_spread_middle>0]/bitstamp_sliced.loc['2018-02-24']
spread_return_middle.head()
```

```python
#Calculate arbitrage spread on March 26, 2018
arbitrage_spread_late = bitstamp_sliced.loc['2018-03-26']-coinbase_sliced.loc['2018-03-26']
arbitrage_spread_late.describe()
```

```python
#find spread greater than zero on March 26
spread_return_late=arbitrage_spread_late[arbitrage_spread_late>0]/coinbase_sliced.loc['2018-03-26']
spread_return_late.head()
```

```python
#find profitable trades for Jan 15
profitable_trades_early=spread_return_early[spread_return_early>.01]
profitable_trades_early.head()
```

```python
#find profitable trades for Feb 24
profitable_trades_middle=spread_return_middle[spread_return_middle > .01]
profitable_trades_middle.head()
```

```python
#find profitable trades for March 26
profitable_trades_late=spread_return_late[spread_return_late>.01]
profitable_trades_late.head()
```

```python
profitable_trades_early.describe()
```

```python
#calulate profit for early
profit_early = profitable_trades_early * coinbase_sliced.loc['2018-01-15']
profit_early.describe()
```

```python
#calculate profit for middle
profit_middle = profitable_trades_middle * bitstamp_sliced.loc['2018-02-24']
profit_middle.describe
```

```python
#finding profitable late 
profit_late=profitable_trades_late * bitstamp_sliced.loc['2018-03-26']
profit_late.describe()
```

```python
profit_per_trade_early = profit_early.dropna()
profit_per_trade_early.describe()
```

```python
profit_per_trade_middle = profit_middle.dropna()
profit_per_trade_middle.describe()
```

```python
profit_per_trade_late = profit_late.dropna()
profit_per_trade_late.describe()
```

```python
#view early profit df
profit_per_trade_early.describe()
```

```python
profit_per_trade_early.plot(figsize=(10,7), title="Profit Per Trade, Jan 15 2018")
```

```python
#calculate the sum profits
profit_per_trade_early.sum()
```

```python
#calculate the early profit
cumulative_profit_early = profit_per_trade_early.cumsum()
cumulative_profit_early
```
